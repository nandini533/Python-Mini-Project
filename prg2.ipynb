```python
def process_weather_column(file_path):
    try:
        # Read the CSV file
        df = pd.read_csv("amazon_delivery.csv")
        
        # Check if 'Weather' column exists
        if 'Weather' in df.columns:
            # Replace spaces with underscores in 'Weather' column
            df['Weather'] = df['Weather'].astype(str).str.replace(' ', '_')
            
            # Save the modified file
            output_path = file_path.replace('.csv', '_modified.csv')
            df.to_csv(output_path, index=False)
            print(f"Processed file saved as: {output_path}")
        else:
            print("Column 'Weather' not found in the CSV file.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Provide the file path
file_path = "amazon_delivery.csv"
process_weather_column(file_path)
```

```python
# Load the dataset
df = pd.read_csv("amazon_delivery.csv")

# Ensure column names are correctly formatted (trim spaces, standardize case)
df.columns = df.columns.str.strip()

# Check if required columns exist to avoid KeyErrors
required_columns = {"Weather", "Area", "Agent_Age"}
if not required_columns.issubset(df.columns):
    raise KeyError(f"Missing required columns: {required_columns - set(df.columns)}")

# Convert 'Agent_Age' to numeric (handle errors to avoid issues)
df["Agent_Age"] = pd.to_numeric(df["Agent_Age"], errors="coerce")

# Group by 'Weather' and 'Area', then calculate the average 'Agent_Age'
result = df.groupby(["Weather", "Area"])["Agent_Age"].mean().reset_index()

# Display the result
print(result)

# Save the output to a new CSV file if needed
result.to_csv("average_agent_age_by_weather_and_area.csv", index=False)
```

```python
df = pd.read_csv("amazon_delivery.csv")

# Group by 'Category' and compute min and max for numerical columns
grouped_df = df.groupby("Category").agg({
    "Agent_Age": ["min", "max"],
    "Agent_Rating": ["min", "max"],
    "Delivery_Time": ["min", "max"]
}).reset_index()

# Display the result
print(grouped_df)
```

```python
df = pd.read_csv("amazon_delivery.csv")

# Define a function to calculate the range (max - min)
def age_range(series):
    return series.max() - series.min()

# Group by Category and Area, then apply the custom aggregation
result = df.groupby(["Category", "Area"])['Agent_Age'].agg(Age_Range=age_range).reset_index()

# Display result
print(result)
```

```python
df = pd.read_csv("amazon_delivery.csv")

# Group by 'Area' and calculate standard deviation and variance of 'Delivery_Time'
result = df.groupby("Area")["Delivery_Time"].agg(["std", "var"])

# Display the result
print(result)
```
